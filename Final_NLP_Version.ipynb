**Getting and cleaning the car reviews from 2007**




```python
def cleaningFu(fileString):
  lines = fileString.readlines()
  linesList = list ()

  for line in lines:
    #print(line)
    linesList.append(line)

  listOfAuthers=list()
  listOfText=list()

  for text in linesList:
    if (text.startswith('<AUTHOR>')):
      text=text.replace('<AUTHOR>','')
      text=text.replace('</AUTHOR>\n','')
      listOfAuthers.append(text)
    if (text.startswith('<TEXT>')):
      text=text.replace('<TEXT>','')
      text=text.replace('</TEXT>\n','')
      #print(type(text))
      listOfText.append(text)

    
  res = dict(zip(listOfAuthers, listOfText))
  return res
```

**Creating the combiend model for classification**

```python
from google.colab import files
def cleanFoodReviews (fileString):
  lines = fileString.readlines()
  linesList = list ()
  listOfValue=list()
  listOfAuthers=list()
  listOfText=list()
  
  for line in lines:
    #print(line)
    linesList.append(line)
  

  for text in linesList:
    if (text.startswith('review/userId: ')):
      text=text.replace('review/userId: ','')
      text=text.replace('\n','')
      listOfAuthers.append(text)
    
    if (text.startswith('review/text: ')):
      text=text.replace('review/text: ','')
      text=text.replace('\n','')
      listOfText.append(text)
    
    if (text.startswith('review/score: ')):
      text=text.replace('review/score: ','')
      text=text.replace('\n','')
      listOfValue.append(text)
  
  
  res = dict(zip(listOfText, listOfValue))
  return res

def plot_loss(history):
  plt.plot(history.history['loss'], label='loss')
  plt.plot(history.history['val_loss'], label='val_loss')
  plt.ylim([0.6, 0.8])
  plt.xlabel('Epoch')
  plt.ylabel('Error [MPG]')
  plt.legend()
  plt.savefig('FirstGraph.png')
  files.download('FirstGraph.png') 
  plt.grid(True)
```

```python
file1= open("/content/drive/My Drive/NLP/foods.txt",'r', encoding="ISO-8859-1")
dictionaryOfFoodReviews=cleanFoodReviews(file1)
for i in dictionaryOfFoodReviews:
  if (float(dictionaryOfFoodReviews[i])>= 3):
    dictionaryOfFoodReviews[i]='1'
  else:
    dictionaryOfFoodReviews[i]='0'  
```

```python
odel='small_bert/bert_en_uncased_L-4_H-512_A-8' 
down='https://tfhub.dev/tensorflow/small_bert/bert_en_uncased_L-4_H-512_A-8/1'
down2='https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3'
bert_preprocess_model = hub.KerasLayer(down)
tfhub_handle_encoder = down
tfhub_handle_preprocess = down2
```

```python
foodDataframe=pd.DataFrame(list(dictionaryOfFoodReviews.items()),columns=["text","label"])
foodDataframe = foodDataframe.astype({"text": str, "label": np.int32})
textualFeatures=foodDataframe['text']
target=foodDataframe['label']
numeric_dict_ds = tf.data.Dataset.from_tensor_slices((textualFeatures, target))

url = 'https://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz'

dataset = tf.keras.utils.get_file('aclImdb_v1.tar.gz', url,
                                  untar=True, cache_dir='.',
                                  cache_subdir='')

dataset_dir = os.path.join(os.path.dirname(dataset), 'aclImdb')

train_dir = os.path.join(dataset_dir, 'train')


remove_dir = os.path.join(train_dir, 'unsup')
shutil.rmtree(remove_dir)
AUTOTUNE = tf.data.AUTOTUNE
batch_size = 128
seed = 42

raw_train_ds = tf.keras.preprocessing.text_dataset_from_directory(
    'aclImdb/train',
    
    validation_split=0.2,
    subset='training',
    seed=seed)

class_names = raw_train_ds.class_names
train_ds = raw_train_ds.cache().prefetch(buffer_size=AUTOTUNE)

val_ds = tf.keras.preprocessing.text_dataset_from_directory(
    'aclImdb/train',
    
    validation_split=0.2,
    subset='validation',
    seed=seed)

val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)

test_ds = tf.keras.preprocessing.text_dataset_from_directory(
    'aclImdb/test',
   )

test_ds = test_ds.cache().prefetch(buffer_size=AUTOTUNE)
test_ds=test_ds.unbatch()

train_ds=train_ds.unbatch()

val_ds=val_ds.unbatch()

ds = train_ds.concatenate(test_ds).concatenate(val_ds)
ds=ds.concatenate(numeric_dict_ds)
train_size = int(75000 * 0.7)
val_size = int(75000 * 0.1)
test_size = int(75000*0.15)


full_dataset = ds.shuffle(buffer_size=100)
full_dataset = ds.take(75000)
train_dataset = full_dataset.take(int(train_size))
train_dataset=train_dataset.batch(int(batch_size))
test_dataset = full_dataset.skip(int(train_size))
test_dataset=test_dataset.batch(int(batch_size))
val_dataset = full_dataset.skip(int(train_size+test_size))
val_dataset=val_dataset.batch(int(batch_size))
#test_dataset = test_dataset.take(int(test_size))


del test_ds
del train_ds
del val_ds
del raw_train_ds

del dataset
del full_dataset
```

```python
bert_preprocess_model = hub.KerasLayer(tfhub_handle_preprocess)
```

```python
bert_model = hub.KerasLayer(tfhub_handle_encoder)

```

```python
def build_classifier_model():
  text_input = tf.keras.layers.Input(shape=(), dtype=tf.string, name='text')
  preprocessing_layer = hub.KerasLayer(tfhub_handle_preprocess, name='preprocessing')
  encoder_inputs = preprocessing_layer(text_input)
  encoder = hub.KerasLayer(tfhub_handle_encoder, trainable=True, name='BERT_encoder')
  outputs = encoder(encoder_inputs)
  net = outputs['pooled_output']
  net = tf.keras.layers.Dropout(0.1)(net)
  net = tf.keras.layers.Dense(1, activation="sigmoid", name='classifier')(net)
  return tf.keras.Model(text_input, net)

classifier_model = build_classifier_model()

```

```python
loss = tf.keras.losses.BinaryCrossentropy(from_logits=True)
metrics = tf.metrics.BinaryAccuracy()
epochs = 5
steps_per_epoch =5 #tf.data.experimental.cardinality(train_dataset).numpy()
print(steps_per_epoch)
num_train_steps =100 #steps_per_epoch * epochs
print(num_train_steps)
num_warmup_steps =int(0.1*num_train_steps)
print(num_warmup_steps)
init_lr = 3e-5
optimizer = optimization.create_optimizer(init_lr=init_lr,
                                          num_train_steps=num_train_steps,
                                          num_warmup_steps=num_warmup_steps,
                                          
                                          optimizer_type='adamw')
classifier_model.compile(optimizer=optimizer,
                         loss=loss,
                         metrics=metrics)
print(f'Training model with {tfhub_handle_encoder}')
history = classifier_model.fit(x=train_dataset,
                               
                               
                               use_multiprocessing= False,
                               validation_data=val_dataset,
                               epochs=epochs)
plot(history)
```

```python
loss, accuracy = classifier_model.evaluate(test_dataset)

print(f'Loss: {loss}')
print(f'Accuracy: {accuracy}')
```
